```python
#%matplotlib inline
# -*- coding:utf-8 -*-
import time
import math
import numpy as np
import pandas as pd
import os
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['STHeiti'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示负号
plt.rc('figure', figsize=(10, 6))
np.set_printoptions(precision=4)
import matplotlib.font_manager as fm
myfont = fm.FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)   

mydata=pd.read_excel('mydata.xlsx')
```

```python
mydata
```

```python
x = list(mydata["region"])
y = list(mydata["per_patent"])

plt.plot(list(mydata['per_patent']),"o--",label = u"各省人均专利数（件/万人）")
plt.xticks(range(len(x)),x,fontproperties = myfont,rotation = 90)

plt.xlabel(u"省份",fontproperties = myfont)
plt.ylabel(u"各省人均专利数（件/万人）",fontproperties = myfont)

plt.legend(loc=2, prop={'family':'SimHei','size':13})

plt.twinx()
plt.plot(list(mydata['patent_case']),"x--",color = 'r',label = u"专利案件数（件）")
plt.xticks(range(len(x)),x,fontproperties = myfont,rotation = 90)

plt.ylabel(u"专利案件数（件）",fontproperties = myfont)

plt.legend(loc=0, prop={'family':'SimHei','size':13})


plt.show()

plt.savefig("test.jpg")

```
